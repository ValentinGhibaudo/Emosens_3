```python
import numpy as np
import matplotlib.pyplot as plt
import ghibtools as gh
import pandas as pd
import seaborn as sns
import pingouin as pg
from params import *
from bibliotheque import processing_raw_maia, processing_stai_longform, processing_short_stai
```

```python
subject_keys = ['P01','P02','P03','P04','P05','P06','P07','P08','P09','P10','P11','P12','P13','P14','P15','P16','P17','P18','P19','P20','P21','P23','P24','P25','P26','P27','P28']
```

```python
concat = []
for participant in subject_keys:
    concat.append(processing_raw_maia(participant))
maias = pd.concat(concat)
```

```python
maias
```

```python
maia_metrics = ['noticing', 'not_distracting', 'not_worrying',
       'attention_regulation', 'emotional_awareness', 'self_regulation',
       'body_listening', 'trusting', 'awareness_of_body_sensations',
       'emotional_reaction', 'capicity_regulation_attention',
       'awareness_of_mind_body', 'trusting_body_sensations', 'global_mean']
```

```python
nrows = 2
ncols = 7
maia_metrics_grid = np.array(maia_metrics).reshape(nrows, ncols)
fig, axs = plt.subplots(nrows, ncols, figsize = (25,8))

for r in range(nrows):
    for c in range(ncols):
        ax = axs[r,c]
        metric = maia_metrics_grid[r,c]
        ax.hist(maias[metric], bins = 10)
        ax.set_title(metric)
        ax.axvline(maias.set_index('participant').loc['P31',metric], color ='red')
plt.savefig('../Figures/Tofile/maia')
plt.show()                                                                             
```

```python
fig, axs = plt.subplots(nrows = 2, ncols = 2 , figsize = (15,10), constrained_layout = True)

for col, predictor in enumerate(['odeur','bloc']):
    if predictor == 'odeur':
        order = ['odor_o','odor_+','odor_-']
    elif predictor == 'bloc':
        order = ['free','lent','confort','rapide']
    for row, outcome in enumerate(['relax','tat']):
        ax = axs[row, col]
        gh.auto_stats(df=short_stais, predictor = predictor , outcome = outcome, subject = 'participant', order = order, ax=ax)
```
