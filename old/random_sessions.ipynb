```python
import random
import pandas as pd
import os
```

```python
sessions = ['music','odor']
```

```python
rows = []
for i in range(1, 41):
    if i < 10:
        sub = f'P0{i}'
    else:
        sub = f'P{i}'
    random_sessions = sessions.copy()
    random.shuffle(random_sessions)
    rows.append([sub, random_sessions[0], random_sessions[1]])
```

```python
random_df = pd.DataFrame(rows, columns = ['participant','stim1','stim2'])
```
